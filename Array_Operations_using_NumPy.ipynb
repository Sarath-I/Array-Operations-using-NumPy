# **Array Operations using NumPy**

## **1. Import numpy as np and see the version (Score 1)**

```python
import numpy as np
print("NumPy Version:", np.__version__)
```

## **2. Create a numpy array containing the numbers from 1 to 10, and then reshape it to a 2x5 matrix. (Score:1)**

```python
arr = np.arange(1, 11)
reshaped_arr = arr.reshape(2, 5)

print("Array:", arr)
print("\nReshaped Matrix:\n", reshaped_arr)
```

## **3. Create a numpy array containing the numbers from 1 to 20, and then extract the elements between the 5th and 15th index. (Score :2)**

```python
arr = np.arange(1, 21)
extract_elements = arr[5:16]

print("Array", arr)
print("\nElements between 5th and 15th index:", extract_elements)
```

## **4. How to compute the mean, median, standard deviation of a numpy array? Use the array you created above (Score 3)**

```python
mean = np.mean(arr)
median = np.median(arr)
std_deviation = np.std(arr)

print("Array:", arr)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_deviation)
```

### **5. Write a NumPy program that creates a 2D array x of shape (3, 4) and a 1D array y of shape (4,). Subtract y from each row of x using broadcasting. (Score : 2) Note : Score 1 for timely submission.**

```python
x = np.array([[5, 10, 15, 20],
              [25, 30, 35, 40],
              [45, 50, 55, 60]])

y = np.array([1, 2, 3, 4])

result = x - y

print("x (2D Array):\n", x)
print("\n y (1D Array):", y)
print("\nResult (x - y):\n", result)
```
